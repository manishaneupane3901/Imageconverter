```python
# importing computer Vision cv library and matplotlib

import cv2 as cv
from matplotlib import pyplot as pt
# loading the image

image = cv.imread('test.jpg')
```

```python
def Showimage(title='',image=None,size=10):
  h,w = image.shape[0],image.shape[1]
  aspect_ratio = w/h
  pt.figure(figsize=(size * aspect_ratio, size))
  pt.title(title)
  pt.imshow(cv.cvtColor(image,cv.COLOR_BGR2RGB))
```

```python
Showimage("Original Image",image,6)
```

```python
new_image = cv.cvtColor(image,cv.COLOR_BGR2GRAY)
Showimage("Converted image",new_image,6)
```
